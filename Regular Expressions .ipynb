### Question 1- Write a Python program to replace all occurrences of a space, comma, or dot with a colon.


```python
def replacecharacter(text):
    replacements = {' ': ':', ',': ':', '.': ':'}
    for oldcharacter, newcharacter in replacements.items():
        text = text.replace(oldcharacter, newcharacter)
    return text

sampletext = 'Python Exercises, PHP exercises.'
newtext = replacecharacter(sampletext)
print(newtext)
```

### Question 5- Create a function in Python to remove the parenthesis in a list of strings. The use of the re.compile() method is mandatory.



```python
import re

def removeparentheses(strings):
    pattern = re.compile(r'[\(\)]')
    processedstrings = [pattern.sub('', s) for s in strings]
    return processedstrings

sampletext = ["example (.com)", "hr@fliprobo (.com)", "github (.com)", "Hello (Data Science World)", "Data (Scientist)"]
result = removeparentheses(sampletext)

for processedstring in result:
    print(processedstring)
```

### Question 6- Write a python program to remove the parenthesis area from the text stored in the text file using Regular Expression.



```python
import re

def removeparentheses(strings):
    pattern = re.compile(r'\s*\([^)]*\)')
    processedstrings = [pattern.sub('', s) for s in strings]
    return processedstrings

sampletext = ["example (.com)", "hr@fliprobo (.com)", "github (.com)", "Hello (Data Science World)", "Data (Scientist)"]
result = removeparentheses(sampletext)

for processedstring in result:
    print(processedstring)
```

### Question 8- Create a function in python to insert spaces between words starting with numbers.



```python
import re

def insertspaces(text):
    modifiedtext = re.sub(r'(?<=\d)(?=[A-Za-z])', ' ', text)
    modifiedtext = re.sub(r'(?<=[A-Za-z])(?=\d)', ' ', modifiedtext)
    return modifiedtext

sampletext = "RegularExpression1IsAn2ImportantTopic3InPython"
result = insertspaces(sampletext)
print(result)
```

### Question 10- Write a python program to extract email address from the text stored in the text file using Regular Expression.

```python
import re

def extractemailsfromtext(text):
    emailpattern = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,7}\b'
    emails = re.findall(emailpattern, text)
    return emails

filename = "sampletext.txt" 

with open(filename, 'r') as file:
    sampletext = file.read()
    extractedemails = extractemailsfromtext(sampletext)
    for email in extractedemails:
        print(email)
```

### Question 11- Write a Python program to match a string that contains only upper and lowercase letters, numbers, and underscores.

```python
import re

def validstring(inputstring):
    pattern = re.compile(r'^[A-Za-z0-9_]*$')
    return bool(pattern.match(inputstring))

teststrings = ["Hello123", "Test_String", "abc_123", "Invalid@String", "123"]
for string in teststrings:
    isvalid = validstring(string)
    if isvalid:
        print(f"'{string}' is a valid string.")
    else:
        print(f"'{string}' is not a valid string.")
```

### Question 13- Write a Python program to remove leading zeros from an IP address

```python
def leadingzeros(ip_address):
    components = ip_address.split('.')
    cleanedcomponents = [str(int(component)) for component in components]
    cleanedip = '.'.join(cleanedcomponents)
    return cleanedip

ipaddresswithzeros = "192.168.001.001"
cleanedipaddress = leadingzeros(ipaddresswithzeros)
print(f"Original IP: {ipaddresswithzeros}")
print(f"Cleaned IP: {cleanedipaddress}")
```

### Question 15- Write a Python program to search some literals strings in a string. 

```python
def literalstring(sampletext, searchwords):
    foundstrings = []
    for search_wrd in searchwords:
        if search_wrd in sampletext:
            foundstrings.append(search_wrd)
    return foundstrings

sampletext = "The quick brown fox jumps over the lazy dog"
searchwords = ["fox", "dog", "horse"]
foundstrings = literalstring(sampletext, searchwords)

if foundstrings:
    print("Found the following strings:")
    for found_str in foundstrings:
        print(found_str)
else:
    print("No matching strings found.")

```

### Question 16- Write a Python program to search a literals string in a string and also find the location within the original string where the pattern occurs



```python
def findlocation(sampletext, searchword):
    foundlocations = []
    start = 0
    while True:
        location = sampletext.find(searchword, start)
        if location == -1:
            break
        foundlocations.append(location)
        start = location + 1
    return foundlocations

sampletext = "The quick brown fox jumps over the lazy dog."
searchword = "fox"
foundlocations = findlocation(sampletext, searchword)

if foundlocations:
    print(f"Found '{searchword}' at the following locations:")
    for location in foundlocations:
        print(location)
else:
    print(f"'{searchword}' not found in the text.")
```

### Question 17- Write a Python program to find the substrings within a string.



```python
def substrings(sampletext, pattern):
    foundpositions = []
    start = 0
    while True:
        position = sampletext.find(pattern, start)
        if position == -1:
            break
        foundpositions.append(position)
        start = position + 1
    return foundpositions

sampletext = 'Python exercises, PHP exercises, C# exercises'
pattern = 'exercises'
foundpositions = substrings(sampletext, pattern)

if foundpositions:
    print(f"Found '{pattern}' at the following positions:")
    for position in foundpositions:
        print(position)
else:
    print(f"'{pattern}' not found in the text.")
```

### Question 18- Write a Python program to find the occurrence and position of the substrings within a string.

```python
def findsubstrings(sampletext, pattern):
    occurrences = []
    start = 0
    while True:
        position = sampletext.find(pattern, start)
        if position == -1:
            break
        occurrences.append(position)
        start = position + 1
    return occurrences

def main():
    sampletext = 'Python exercises, PHP exercises, C# exercises'
    pattern = 'exercises'
    occurrences = findsubstrings(sampletext, pattern)

    if occurrences:
        print(f"Found '{pattern}' at the following positions:")
        for index, position in enumerate(occurrences, start=1):
            print(f"Occurrence {index}: Position {position}")
    else:
        print(f"'{pattern}' not found in the text.")

if __name__ == "__main__":
    main()
```

### Question 19- Write a Python program to convert a date of yyyy-mm-dd format to dd-mm-yyyy format.

```python
def convert_date_format(input_date):
    parts = input_date.split('-')
    if len(parts) == 3:
        year, month, day = parts
        converted_date = f"{day}-{month}-{year}"
        return converted_date
    else:
        return "Invalid date format"

date_yyyy_mm_dd = "2020-03-20"
converted_date_dd_mm_yyyy = convert_date_format(date_yyyy_mm_dd)
print(f"Original date: {date_yyyy_mm_dd}")
print(f"Converted date: {converted_date_dd_mm_yyyy}")
```

### Question 21- Write a Python program to separate and print the numbers and their position of a given string.

```python
def separate_numbers_and_positions(samplestring):
    numbers = []
    positions = []
    for position, char in enumerate(samplestring):
        if char.isdigit():
            numbers.append(char)
            positions.append(position)
    return numbers, positions

samplestring = "Hello123World4567!"
numbers, positions = separate_numbers_and_positions(samplestring)

print("Numbers found:", numbers)
print("Positions:", positions)
```

### Question 22- Write a regular expression in python program to extract maximum/largest numeric value from a string.



```python
import re

def max_numeric_value(text):
    numbers = re.findall(r'\d+', text)
    if numbers:
        max_number = max(map(int, numbers))
        return str(max_number)
    else:
        return "No numeric values found"

sampletext = 'My marks in each semester are: 947, 896, 926, 524, 734, 950, 642'
max_value = max_numeric_value(sampletext)
print("Maximum numeric value is", max_value)
```

### Question 25- Write a Python program to remove continuous duplicate words from Sentence using Regular Expression.


```python
import re

def continuous_duplicates(sampletext):
    pattern = r'\b(\w+)\s+\1\b'
    cleanedsentence = re.sub(pattern, r'\1', sampletext, flags=re.IGNORECASE)
    return cleanedsentence

sampletext = "Hello hello world world"
cleanedtext = continuous_duplicates(sampletext)
print(cleanedtext)
```

### Question 26-  Write a python program using RegEx to accept string ending with alphanumeric character.

```python
import re

def ending_with_alphanumeric(inputstring):
    pattern = r'^.*[A-Za-z0-9]$'
    return bool(re.match(pattern, inputstring))

samplestrings = ["Hello123", "abc_", "Test123_", "Python2023"]
for string in samplestrings:
    if ending_with_alphanumeric(string):
        print(f"'{string}' ends with an alphanumeric character.")
    else:
        print(f"'{string}' does not end with an alphanumeric character.")
```

### Question 28- Write a python program using RegEx to remove <U+..> like symbols


```python
import re

def remove_u_symbols(sampletext):
    pattern = r'<U\+\w{2,4}>'
    cleanedtext = re.sub(pattern, '', sampletext)
    return cleanedtext

sampletext = "@Jags123456 Bharat band on 28??<ed><U+00A0><U+00BD><ed><U+00B8><U+0082>Those who  are protesting #demonetization  are all different party leaders"
cleanedtext = remove_u_symbols(sampletext)
print(cleanedtext)
```

### Question 29- Write a python program to extract dates from the text stored in the text file.



```python
import re

def extract_dates_from_file(filename):
    with open(filename, 'r') as file:
        content = file.read()
        dates = re.findall(r'\d{2}-\d{2}-\d{4}', content)
        return dates

file_name = "sampletext3.txt"
extracted_dates = extract_dates_from_file(file_name)

if extracted_dates:
    print("Extracted dates:", extracted_dates)
else:
    print("No dates found in the text.")
```

### Question 30- Create a function in python to remove all words from a string of length between 2 and 4.
### The use of the re.compile() method is mandatory.

```python
import re

def words_of_length_between_2_and_4(sampletext):
    pattern = re.compile(r'\b\w{2,4}\b')
    cleanedtext = pattern.sub('', sampletext)
    return cleanedtext

sampletext = "The following example creates an ArrayList with a capacity of 50 elements. 4 elements are then added to the ArrayList and the ArrayList is trimmed accordingly."
cleanedtext = words_of_length_between_2_and_4(sampletext)
print(cleanedtext)
```
